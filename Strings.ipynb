```python
# How to add a ' or "" in a string

course = 'Python Course for beginners'
print(course)
course = "Python's Course for beginners"
print(course)
course = 'Python for "beginners"'
print(course)
```

```python
# Adding multiple lines

email = '''
Hi John,

This is our first email to you.

Thank you,
Dilni

'''

print(email)
```

```python
# How to get characters from strings

course = 'Python for beginners'

print('1.')
print(course[0])
print('2.')
print(course[-1])
print('3.')
print(course[0:5])
print('4.')
print(course[1:-1])
print('5.')
print(course[-5:-1])
print('6.')
print(course[0:])
print('7.')
print(course[:5])
print('8.')
print(course[:])
```

## Formatting Strings

```python
# Unformatted

first = 'John'
last = 'Smith'

message = first + ' [' + last + '] is a coder.'
print(message)
```

```python
# As the code gets more complex, it gets harder to visualize these strings like this
#Therefore

# Formatted Strings

msg = f'{first} [{last}] is a coder.'

print(msg)
```

## String Methods
Really cool things we can do with strings

```python
course = 'Python for Beginners'

# length function
print(len(course))

print()

# Upper and Lower Case
print(course.upper())
print(course.lower())
print(course.capitalize())

print()

# Find and replace
print(course.find('P'))
print(course.find('yth'))
print(course.replace('Beginners', 'Absolute Beginners'))

print()
print('Python' in course)
print('python' in course)
```
