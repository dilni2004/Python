## Exercise

Ask for their name and favourite colour. Then print it as a sentence

```python
name = input('What is your name? ')
fav_colour = input('What is your favourite colour? ')

print(name + ' likes ' + fav_colour + ' colour')
```

Ask for the year user was born in and calculate their age.

```python
import datetime

birth_year = input('Birth Year: ')
print(type(birth_year))

# age = 2025 - birth_year
# or you can get the current year from the code itself

age = datetime.datetime.now().year - int(birth_year)
print(type(age))

print('age : ', age)
```
